## 前処理0. horse_result_tableの整形

```python
# idのカラムをすべて文字列型に変換し、前後の空白を削除する
df_race["race_id"] = df_race["race_id"].astype(str).str.strip()
df_race["horse_id"] = df_race["horse_id"].astype(str).str.strip()
df_horse["race_id"] = df_horse["race_id"].astype(str).str.strip()
df_horse["horse_id"] = df_horse["horse_id"].astype(str).str.strip()

# jockey_idとtrainer_idも同様に処理する
# スクレイピングをしてtableを作成する際にこの処理はやっておく
```

```python
# 型をあらかじめ揃えておいたときに使う方のコード

# what: horse_result_tableの前処理
# for:  horse_result_tableをrace_result_tableに1対1でマージできるように
# how:  horse_result_tableのrace_idを期間で絞り込み、出走頭数と一致するデータのみを残す。さらに地方競馬場のデータを除外

# 直近nレースから集約できるように並び替えておく
df_horse = df_horse.sort_values(by=["horse_id", "race_date"], ascending=[True, True])
# race_idをスクレイピングした期間のデータに絞る
df_horse_indate = df_horse[(df_horse["race_date"] >= "2023-01-01") & (df_horse["race_date"] <= "2024-04-30")].copy()

# 1. "取"(取消), "除"(除外), ""(空欄)になっている行を出走頭数(actual_horses)のカウントから除外する
df_horse_indate.loc[:, "is_cash"] = df_horse_indate["finish_position"].astype(str).str.strip().isin(["取", "除", ""])
actual_counts = df_horse_indate[~df_horse_indate["is_cash"]].groupby("race_id").size().rename("actual_horses")

# 2. num_of_horses と結合
df_check = df_horse_indate.merge(actual_counts, on="race_id")
# 2.5 num_of_horses を数値型に変換
df_check.loc[:, "num_of_horses"] = pd.to_numeric(df_check["num_of_horses"], errors="coerce")

# 3. 一致しているデータだけ残す
df_clean = df_check[df_check["actual_horses"] == df_check["num_of_horses"]].copy()

# 4. 地方競馬場のデータを除外する
# race_idの5-6桁目を抽出（1-basedで5-6桁 → 0-based slice [4:6]）
df_clean.loc[:, 'race_course_code'] = df_clean['race_id'].str[4:6]
jra_codes = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10"]  # JRAの中央競馬場コード
df_clean = df_clean[df_clean['race_course_code'].isin(jra_codes)]

# num_of_horsesとrace_course_codeを数値型に変換
df_clean["num_of_horses"] = pd.to_numeric(df_clean["num_of_horses"], errors="coerce")
df_clean["race_course_code"] = pd.to_numeric(df_clean["race_course_code"], errors="coerce")
df_clean
```

```python
# what: horse_result_tableの前処理
# for:  horse_result_tableをrace_result_tableに1対1でマージできるように
# how:  horse_result_tableのrace_idを期間で絞り込み、出走頭数と一致するデータのみを残す。さらに地方競馬場のデータを除外

# 直近nレースから集約できるように並び替えておく
df_horse = df_horse.sort_values(by=["horse_id", "race_date"], ascending=[True, True])
# race_idをスクレイピングした期間のデータに絞る
df_horse_indate = df_horse[(df_horse["race_date"] >= "2023-10-01") & (df_horse["race_date"] <= "2024-04-30")].copy()

# 1. "取"(取消), "除"(除外), ""(空欄)になっている行を出走頭数(actual_horses)のカウントから除外する
df_horse_indate.loc[:, "is_cash"] = df_horse_indate["finish_position"].astype(str).str.strip().isin(["取", "除", ""])
actual_counts = df_horse_indate[~df_horse_indate["is_cash"]].groupby("race_id").size().rename("actual_horses")

# 2. num_of_horses と結合
df_check = df_horse_indate.merge(actual_counts, on="race_id")
# 2.5 num_of_horses を数値型に変換
df_check.loc[:, "num_of_horses"] = pd.to_numeric(df_check["num_of_horses"], errors="coerce")

# 3. 一致しているデータだけ残す
df_clean = df_check[df_check["actual_horses"] == df_check["num_of_horses"]].copy()

# 4. 地方競馬場のデータを除外する
# race_idの5-6桁目を抽出（1-basedで5-6桁 → 0-based slice [4:6]）
df_clean.loc[:, 'race_course_code'] = df_clean['race_id'].str[4:6]
jra_codes = ["01", "02", "03", "04", "05", "06", "07", "08", "09", "10"]  # JRAの中央競馬場コード
df_clean = df_clean[df_clean['race_course_code'].isin(jra_codes)]

# num_of_horsesとrace_course_codeを数値型に変換
df_clean["num_of_horses"] = pd.to_numeric(df_clean["num_of_horses"], errors="coerce")
df_clean["race_course_code"] = pd.to_numeric(df_clean["race_course_code"], errors="coerce")
df_clean
```

## 前処理1. 対数化したオッズの取得

```python
# オッズがstr型だったため数値に変更
# df_clean["odds"] = pd.to_numeric(df_clean["odds"], errors="coerce")
# 対数化
df_clean['log_odds'] = np.log(df_clean['odds'])
```

```python
# 馬ごと各レース時点での過去3, 5, 10試合のlog_oddsの平均をとる

# 「今見ている行より下側」の log_odds を参照するために shift(-1)
df_clean["log_odds_shift"] = df_clean.groupby("horse_id")["log_odds"].shift(+1)
# レース時点での過去3, 5, 10レースの平均を計算
df_clean["log_odds_avg3"] = (
    df_clean.groupby("horse_id")["log_odds_shift"]
      .rolling(3)
      .mean()
      .reset_index(level=0, drop=True)
)

df_clean["log_odds_avg5"] = (  
    df_clean.groupby("horse_id")["log_odds_shift"]
      .rolling(5)
      .mean()
      .reset_index(level=0, drop=True)
)

df_clean["log_odds_avg10"] = (
    df_clean.groupby("horse_id")["log_odds_shift"]
      .rolling(10)
      .mean()
      .reset_index(level=0, drop=True)
)
# 不要な列を削除
df_clean = df_clean.drop(columns=["log_odds_shift"])
```

## 前処理2. 平均との差分を分布で割ったnレース平均(z-score)

```python
df_clean['log_odds_zscore'] = (
    df_clean.groupby('race_id')['log_odds']
      .transform(lambda x: (x - x.mean()) / (x.std(ddof=0) + 1e-9))
)

# 「今見ている行より下側」の log_odds を参照するために shift(-1)
df_clean["log_odds_zscore_shift"] = df_clean.groupby("horse_id")["log_odds_zscore"].shift(+1)
# レース時点での過去3, 5, 10レースの平均を計算
df_clean["log_odds_zscore_avg3"] = (
    df_clean.groupby("horse_id")["log_odds_zscore_shift"]
      .rolling(3)
      .mean()
      .reset_index(level=0, drop=True)
)

df_clean["log_odds_zscore_avg5"] = (  
    df_clean.groupby("horse_id")["log_odds_zscore_shift"]
      .rolling(5)
      .mean()
      .reset_index(level=0, drop=True)
)

df_clean["log_odds_zscore_avg10"] = (
    df_clean.groupby("horse_id")["log_odds_zscore_shift"]
      .rolling(10)
      .mean()
      .reset_index(level=0, drop=True)
)
# 不要な列を削除
df_clean = df_clean.drop(columns=["log_odds_zscore_shift"])
```

## 前処理3. 人気(popularity)

```python
# str型だったため数値に変更
df_clean["popularity"] = pd.to_numeric(df_clean["popularity"], errors="coerce")
```

### popularity_inverse（人気の逆数）

```python
df_clean["popularity_inverse"] = 1 / df_clean["popularity"]
df_clean["popularity_inverse_avg3"] = df_clean.groupby("horse_id")["popularity_inverse"].shift(1).rolling(3).mean()
df_clean["popularity_inverse_avg5"] = df_clean.groupby("horse_id")["popularity_inverse"].shift(1).rolling(5).mean()
df_clean["popularity_inverse_avg10"] = df_clean.groupby("horse_id")["popularity_inverse"].shift(1).rolling(10).mean()
```

### popularity_bucket（人気のビン分割）

```python
def popularity_bucket(pop):
    if pop <= 2:
        return 1      # 超上位人気
    elif pop <= 5:
        return 2      # 上位人気
    elif pop <= 9:
        return 3      # 中穴
    else:
        return 4      # 大穴

df_clean['popularity_bucket'] = df_clean['popularity'].apply(popularity_bucket)
df_clean["popularity_bucket_avg3"] = df_clean.groupby("horse_id")["popularity_bucket"].shift(1).rolling(3).mean()
df_clean["popularity_bucket_avg5"] = df_clean.groupby("horse_id")["popularity_bucket"].shift(1).rolling(5).mean()
df_clean["popularity_bucket_avg10"] = df_clean.groupby("horse_id")["popularity_bucket"].shift(1).rolling(10).mean()
```

## 前処理4. 斤量（weight_carried）

### weight_minus_avg_in_race（レース内平均との差斤量）

```python
# df_clean['weight'] = pd.to_numeric(df_clean['weight'], errors='coerce')
df_clean['weight_avg_in_race'] = (
    df_clean['weight_carried'] - df_clean.groupby('race_id')['weight_carried'].transform('mean')
)
```

### weight_per_body_weight（馬体重に対する斤量比）

```python
df_clean['weight_per_body_weight'] = df_clean['weight_carried'] / df_clean['weight_horse']
```

## 前処理5. 最終順位（finish_position）

### finish_position_inverse（最終順位の逆順）

```python
df_clean["finish_position_inverse"] = 1 / df_clean["finish_position"]
df_clean["finish_position_inverse_avg3"] = df_clean.groupby("horse_id")["finish_position_inverse"].shift(1).rolling(3).mean()
df_clean["finish_position_inverse_avg5"] = df_clean.groupby("horse_id")["finish_position_inverse"].shift(1).rolling(5).mean()
df_clean["finish_position_inverse_avg10"] = df_clean.groupby("horse_id")["finish_position_inverse"].shift(1).rolling(10).mean()
```

### expected_vs_actual_gap（期待順位と実順位の乖離）

```python
df_clean['expected_vs_actual_gap'] = df_clean['popularity'] - df_clean['finish_position']
df_clean["expected_vs_actual_gap_avg3"] = df_clean.groupby("horse_id")["expected_vs_actual_gap"].shift(1).rolling(3).mean()
df_clean["expected_vs_actual_gap_avg5"] = df_clean.groupby("horse_id")["expected_vs_actual_gap"].shift(1).rolling(5).mean()
df_clean["expected_vs_actual_gap_avg10"] = df_clean.groupby("horse_id")["expected_vs_actual_gap"].shift(1).rolling(10).mean()
```

## 前処理6. 走破タイム（time）

### time_diff_from_winner（勝ち馬とのタイム差）

```python
df_clean["time_diff_from_winner"] = df_clean["time"] - df_clean.groupby("race_id")["time"].transform("min")
df_clean["time_diff_from_winner_avg3"] = df_clean.groupby("horse_id")["time_diff_from_winner"].shift(1).rolling(3).mean()
df_clean["time_diff_from_winner_avg5"] = df_clean.groupby("horse_id")["time_diff_from_winner"].shift(1).rolling(5).mean()
df_clean["time_diff_from_winner_avg10"] = df_clean.groupby("horse_id")["time_diff_from_winner"].shift(1).rolling(10).mean()
```

### time_zscore_in_race（レース内Zスコア）

```python
df_clean["time_zscore_in_race"] = (df_clean["time"] - df_clean.groupby("race_id")["time"].transform("mean")) / df_clean.groupby("race_id")["time"].transform("std")
df_clean["time_zscore_in_race_avg3"] = df_clean.groupby("horse_id")["time_zscore_in_race"].shift(1).rolling(3).mean()
df_clean["time_zscore_in_race_avg5"] = df_clean.groupby("horse_id")["time_zscore_in_race"].shift(1).rolling(5).mean()
df_clean["time_zscore_in_race_avg10"] = df_clean.groupby("horse_id")["time_zscore_in_race"].shift(1).rolling(10).mean()
```

## 前処理7. 着差（margin）

### margin_to_winner（勝ち馬との差）

```python
df_clean["margin_to_winner"] = df_clean.groupby("race_id")["margin"].cumsum()
df_clean["margin_to_winner_avg3"] = df_clean.groupby("horse_id")["margin_to_winner"].shift(1).rolling(3).mean()
df_clean["margin_to_winner_avg5"] = df_clean.groupby("horse_id")["margin_to_winner"].shift(1).rolling(5).mean()
df_clean["margin_to_winner_avg10"] = df_clean.groupby("horse_id")["margin_to_winner"].shift(1).rolling(10).mean()
```

## 前処理8. 上がり3F（final_3f）

### final_3f_rank（レース内順位）

```python
df_clean["final_3f_rank"] = df_clean.groupby("race_id")["final_3f"].rank(ascending=True)
df_clean["final_3f_rank_avg3"] = df_clean.groupby("horse_id")["final_3f_rank"].shift(1).rolling(3).mean()
df_clean["final_3f_rank_avg5"] = df_clean.groupby("horse_id")["final_3f_rank"].shift(1).rolling(5).mean()
df_clean["final_3f_rank_avg10"] = df_clean.groupby("horse_id")["final_3f_rank"].shift(1).rolling(10).mean()
```

### final_3f_diff_from_best（最速との差）

```python
df_clean["final_3f_diff_from_best"] = df_clean["final_3f"] - df_clean.groupby("race_id")["final_3f"].transform("min")
df_clean["final_3f_diff_avg3"] = df_clean.groupby("horse_id")["final_3f_diff_from_best"].shift(1).rolling(3).mean()
df_clean["final_3f_diff_avg5"] = df_clean.groupby("horse_id")["final_3f_diff_from_best"].shift(1).rolling(5).mean()
df_clean["final_3f_diff_avg10"] = df_clean.groupby("horse_id")["final_3f_diff_from_best"].shift(1).rolling(10).mean()
```

### final_3f_zscore

```python
df_clean["final_3f_zscore"] = (df_clean["final_3f"] - df_clean.groupby("race_id")["final_3f"].transform("mean")) / df_clean.groupby("race_id")["final_3f"].transform("std")
df_clean["final_3f_zscore_avg3"] = df_clean.groupby("horse_id")["final_3f_zscore"].shift(1).rolling(3).mean()
df_clean["final_3f_zscore_avg5"] = df_clean.groupby("horse_id")["final_3f_zscore"].shift(1).rolling(5).mean()
df_clean["final_3f_zscore_avg10"] = df_clean.groupby("horse_id")["final_3f_zscore"].shift(1).rolling(10).mean()
```

## 前処理9. 賞金(prize_money)

### log_prize（賞金の対数変換）

```python
# 対数変換（賞金の歪み対策）
df_clean['log_prize_money'] = np.log1p(df_clean['prize_money'])

# 直近N走平均
df_clean['log_prize_avg3'] = df_clean.groupby('horse_id')['log_prize_money'].shift(1).rolling(3).mean()
df_clean['log_prize_avg5'] = df_clean.groupby('horse_id')['log_prize_money'].shift(1).rolling(5).mean()
df_clean['log_prize_avg10'] = df_clean.groupby('horse_id')['log_prize_money'].shift(1).rolling(10).mean()
```

## 前処理10. 通過順位（passing_1st ~ _4th）   

### passing_{n}_rel（相対位置）

```python
df_clean["passing_1st"] = pd.to_numeric(df_clean["passing_1st"], errors="coerce")
df_clean["passing_2nd"] = pd.to_numeric(df_clean["passing_2nd"], errors="coerce")
df_clean["passing_3rd"] = pd.to_numeric(df_clean["passing_3rd"], errors="coerce")
df_clean["passing_4th"] = pd.to_numeric(df_clean["passing_4th"], errors="coerce")
passing_cols = ['passing_1st','passing_2nd','passing_3rd','passing_4th']

# 頭数で割って相対位置に変換
for c in passing_cols:
    df_clean[c + '_rel'] = (df_clean['num_of_horses'] - df_clean[c]) / df_clean['num_of_horses']
    # 直近N走平均（例：3走）
    df_clean[c + '_rel' + '_avg3'] = df_clean.groupby('horse_id')[c + '_rel'].shift(1).rolling(3).mean()
    df_clean[c + '_rel' + '_avg5'] = df_clean.groupby('horse_id')[c + '_rel'].shift(1).rolling(5).mean()
    df_clean[c + '_rel' + '_avg10'] = df_clean.groupby('horse_id')[c + '_rel'].shift(1).rolling(10).mean()
```
