```python
# idのカラムをすべて文字列型に変換し、前後の空白を削除する
df_race["race_id"] = df_race["race_id"].astype(str).str.strip()
df_race["horse_id"] = df_race["horse_id"].astype(str).str.strip()
df_horse["race_id"] = df_horse["race_id"].astype(str).str.strip()
df_horse["horse_id"] = df_horse["horse_id"].astype(str).str.strip()
```

```python
# 直近nレースから集約できるように並び替えておく
df_horse = df_horse.sort_values(by=["horse_id", "race_date"], ascending=[True, True])
# race_idをスクレイピングした期間のデータに絞る
df_horse_filtered = df_horse[(df_horse["race_date"] >= "2023-10-01") & (df_horse["race_date"] <= "2024-04-30")]

# ---------------------
# 出走頭数と実際の頭数を比較して不整合なデータを削除
# ---------------------

# 1. race_id ごとに行数をカウント
race_counts = df_horse_filtered.groupby("race_id").size().rename("actual_horses")

# 2. num_of_horses と結合
df_check = df_horse_filtered.merge(race_counts, on="race_id")

# 2.5 num_of_horses を数値型に変換
df_check["num_of_horses"] = pd.to_numeric(df_check["num_of_horses"], errors="coerce")

# 3. 一致している race_id だけ残す
df_clean = df_check[df_check["actual_horses"] == df_check["num_of_horses"]]
```

```python
merged = df_race.merge(
    df_clean,
    on=["horse_id", "race_id"],
    how="outer",
    indicator=True
)
matched_count = (merged["_merge"] == "both").sum()
print(f"Number of matched records: {matched_count}")
unmatched_count = (merged["_merge"] != "both").sum()
print(f"Number of unmatched records: {unmatched_count}")
```

### 1. 対数化したオッズの取得

```python
# オッズがstr型だったため数値に変更
df_clean["odds"] = pd.to_numeric(df_clean["odds"], errors="coerce")
# 対数化
df_clean['log_odds'] = np.log(df_clean['odds'])
```

```python
# 馬ごと各レース時点での過去3, 5, 10試合のlog_oddsの平均をとる

# 「今見ている行より下側」の log_odds を参照するために shift(-1)
df_clean["log_odds_shift"] = df_clean.groupby("horse_id")["log_odds"].shift(+1)
# レース時点での過去3, 5, 10レースの平均を計算
df_clean["log_odds_avg_3"] = (
    df_clean.groupby("horse_id")["log_odds_shift"]
      .rolling(3)
      .mean()
      .reset_index(level=0, drop=True)
)

df_clean["log_odds_avg_5"] = (
    df_clean.groupby("horse_id")["log_odds_shift"]
      .rolling(5)
      .mean()
      .reset_index(level=0, drop=True)
)

df_clean["log_odds_avg_10"] = (
    df_clean.groupby("horse_id")["log_odds_shift"]
      .rolling(10)
      .mean()
      .reset_index(level=0, drop=True)
)
# 不要な列を削除
df_clean = df_clean.drop(columns=["log_odds_shift"])

df_clean.head(20)
```

### 2. 平均との差分を分布で割ったnレース平均(z-score)

```python
df_clean['odds_zscore_in_race'] = (
    df_clean.groupby('race_id')['odds']
      .transform(lambda x: (x - x.mean()) / (x.std(ddof=0) + 1e-9))
)
```
